#  Imports and Data

```python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
np.random.seed(42)
```

```python
random_x = np.random.randint(1,101,100)
```

```python
random_y = np.random.randint(1,101,100)
```

# Scatter Plot

```python
data = [go.Scatter(x=random_x,y=random_y,mode='markers',
                  marker=dict(
                      size=12,
                      color='rgb(51,204,153)',
                      symbol='pentagon',
                      line={'width':2}
                      
                  
                  ))]
```

```python
layout = go.Layout(title="Hello First Plot",xaxis={'title':'MY X AXIS'},yaxis=dict(title='MY Y AXIS'),
                   hovermode='closest')
```

```python
fig= go.Figure(data=data,layout=layout)
```

```python
pyo.plot(fig,filename='temp1.html')
```

# Line chart

```python
np.random.seed(56)
```

```python
x_values= np.linspace(0,1,100)
y_values= np.random.randn(100)
```

```python
trace0 = go.Scatter(x=x_values,y=y_values+5,
                  mode='markers',name='markers')
```

```python
trace1 = go.Scatter(x=x_values,y=y_values,
                  mode='lines',name='mylines')
```

```python
trace2 = go.Scatter(x=x_values,y=y_values-5,
                  mode='lines+markers',name='my favorite')
```

```python
data = [trace0,trace1,trace2]
```

```python
layout = go.Layout(title='Line Charts')
```

```python
fig = go.Figure(data=data,layout=layout)
```

```python
pyo.plot(fig)
```

# Bar Charts

```python
import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd
```

```python
df = pd.read_csv('./Data/2018WinterOlympics.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
trace1=go.Bar(x=df['NOC'],y=df['Gold'],name='Gold',marker={'color':'#FFD700'})
```

```python
trace2=go.Bar(x=df['NOC'],y=df['Silver'],name='Silver',marker={'color':'#9EA0A1'})
```

```python
trace3=go.Bar(x=df['NOC'],y=df['Bronze'],name='Bronze',marker={'color':'#CD7F32'})
```

```python
data=[trace1,trace2,trace3]
```
