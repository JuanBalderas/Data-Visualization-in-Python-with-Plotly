#  Imports and Data <a name="imports"></a>

```python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
np.random.seed(42)
```

```python
random_x = np.random.randint(1,101,100)
```

```python
random_y = np.random.randint(1,101,100)
```

# Scatter Plot <a name="scatter"></a>

```python
data = [go.Scatter(x=random_x,y=random_y,mode='markers',
                  marker=dict(
                      size=12,
                      color='rgb(51,204,153)',
                      symbol='pentagon',
                      line={'width':2}
                      
                  
                  ))]
```

```python
layout = go.Layout(title="Hello First Plot",xaxis={'title':'MY X AXIS'},yaxis=dict(title='MY Y AXIS'),
                   hovermode='closest')
```

```python
fig= go.Figure(data=data,layout=layout)
```

```python
pyo.plot(fig,filename='temp1.html')
```

```python
iplot(fig)
```

```python
from plotly import tools
```

```python
# MULTIPLE HEAT MAPS WITH SUBPLOTS
```

```python
dfA = pd.read_csv('./Data/2010SitkaAK.csv')
dfB = pd.read_csv('./Data/2010SantaBarbaraCA.csv')
dfC = pd.read_csv('./Data/2010YumaAZ.csv')
```

```python
trace1 = go.Heatmap(x=dfA['DAY'],y=dfA['LST_TIME'],z=dfA['T_HR_AVG'],colorscale='Jet',zmin=5,zmax=40)
trace2 = go.Heatmap(x=dfA['DAY'],y=dfB['LST_TIME'],z=dfB['T_HR_AVG'],colorscale='Jet',zmin=5,zmax=40)
trace3 = go.Heatmap(x=dfA['DAY'],y=dfC['LST_TIME'],z=dfC['T_HR_AVG'],colorscale='Jet',zmin=5,zmax=40)
```

```python
fig = tools.make_subplots(rows=1,cols=3,subplot_titles=['Sitka AK','SB CA','Yuma AZ'],
                                                       shared_yaxes=False)
```

```python
fig.append_trace(trace1,1,1)
fig.append_trace(trace2,1,2)
fig.append_trace(trace3,1,3)
```

```python
fig['layout'].update(title='Temps for 3 cities')
```

```python
iplot(fig)
```
